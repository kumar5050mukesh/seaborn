```python
# Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
# "timepoint" and y = "signal" for different events and regions.

import seaborn as sns
data=sns.load_dataset("fmri")
data
```

```python
sns.lineplot(x="timepoint",y="signal",data=data,hue="event")
```

```python
sns.lineplot(x="timepoint",y="signal",data=data,hue="region")
```

```python
# Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
# parameter for the 'cut' column of the diamonds dataset.


import seaborn as sns
import matplotlib.pyplot as plt

data_diamonds = sns.load_dataset("diamonds")
sns.displot(data=data_diamonds, x="price", hue="cut", multiple="stack")
plt.show()
```

```python
# Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column
# of the iris dataset.

import seaborn as sns
import matplotlib.pyplot as plt
data_iris=sns.load_dataset("iris")

sns.pairplot(data=data_iris,hue="species")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

flights_data = sns.load_dataset("flights")
print(flights_data)
flights_data = flights_data.pivot("month","year",  "passengers")
print(flights_data)
sns.heatmap(data=flights_data, cmap="viridis")

plt.title("Passenger Numbers on Flights")
plt.xlabel("Year")
plt.ylabel("Month")

plt.show()
```
